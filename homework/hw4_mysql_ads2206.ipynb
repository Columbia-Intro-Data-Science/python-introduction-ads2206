### 2. Make a list of all customer names who have videos out and how much they owe.
``` sql
SELECT c.first_name, c.last_name, count(c.customer_id) AS Num_Vidoes
FROM customer AS c,
     rental AS r
WHERE c.customer_id = r.customer_id
  AND r.return_date IS NULL
GROUP BY c.customer_id
ORDER BY 3 DESC
```

### 4. Using a $1 per day late fee, find out which users owe the most assuming all rentals are a week long. 

```sql
SELECT customer.customer_id,
       customer.first_name,
       customer.last_name,
       date(r.return_date) - date(r.rental_date) - 7 AS late_fee
FROM rental r
INNER JOIN customer ON customer.customer_id = r.customer_id
WHERE date(r.return_date) - date(r.rental_date) - 7 >0
ORDER BY late_fee DESC;
```

### 5.  What hour of the day do people rent the most? 

```sql
SELECT extract(hour FROM rental_date) AS hr_rented, 
		count(extract(hour FROM rental_date)) AS movies_rented_count
FROM rental
GROUP BY hr_rented
ORDER BY movies_rented_count DESC;
```
